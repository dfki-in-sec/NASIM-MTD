```python
#Imports
from os import listdir
from os.path import isfile, join
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}
matplotlib.rc('font', **font)
#matplotlib.rcParams['pdf.fonttype'] = 42
#matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['text.usetex'] = True
plt.rcParams.update({'font.size': 26})
plt.rcParams['figure.figsize'] = (10.0, 7.0)
```

Next get all data from all the files in the result folder:

```python
# Import Result Data
columns = ["agent", "seed", "one_goal", "num_hosts", "num_honeypots", "movement_time", "won", "steps", "unique_success", "total_success"]
path = "results/"
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]
df = pd.concat((pd.read_csv(join(path, f), header=0, names=columns) for f in onlyfiles))
```

```python
print(df.to_string(max_rows=5)) 
#print(df.info)
print(df.describe())
```

## Winning Probability Plot #MT-T

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 7))
# honeypots
ag_hp = df[df['movement_time']=="-"]
ag_hp = ag_hp[ag_hp['agent']=="aggressiv"]
rating_probs = ag_hp.groupby('num_honeypots').size().div(len(ag_hp))
conditional_probs = ag_hp.groupby(['won', 'num_honeypots']).size().div(len(ag_hp)).div(rating_probs, axis=0, level='num_honeypots')
#print(conditional_probs)
conditional_probs.drop(labels=[0]).plot(ax=axes[0], style=['r'], label='aggressive')

ag_hp = df[df['movement_time']=="-"]
ag_hp = ag_hp[ag_hp['agent']=="standard"]
rating_probs = ag_hp.groupby('num_honeypots').size().div(len(ag_hp))
conditional_probs = ag_hp.groupby(['won', 'num_honeypots']).size().div(len(ag_hp)).div(rating_probs, axis=0, level='num_honeypots').drop(labels=[0]).plot(ax=axes[0], style=['b'], label='standard')

ag_hp = df[df['movement_time']=="-"]
ag_hp = ag_hp[ag_hp['agent']=="careful"]
rating_probs = ag_hp.groupby('num_honeypots').size().div(len(ag_hp))
conditional_probs = ag_hp.groupby(['won', 'num_honeypots']).size().div(len(ag_hp)).div(rating_probs, axis=0, level='num_honeypots').drop(labels=[0]).plot(ax=axes[0], style=['g'], label='carefull')

plt.sca(axes[0])
axes[0].title.set_text('a')
plt.xlabel("num_honeypots")
plt.ylabel("probability")
plt.xticks(range(6), [0, 2, 4, 6, 8, 10])

# movement time
ag_hp = df[df['num_honeypots']==0]
ag_hp = ag_hp[ag_hp['agent']=="aggressiv"]
rating_probs = ag_hp.groupby('movement_time').size().div(len(ag_hp))
conditional_probs = ag_hp.groupby(['won', 'movement_time']).size().div(len(ag_hp)).div(rating_probs, axis=0, level='movement_time')
#print(conditional_probs)
conditional_probs.drop(labels=[-1]).plot(ax=axes[1], style=['r'], label='aggressive')

ag_hp = df[df['num_honeypots']==0]
ag_hp = ag_hp[ag_hp['agent']=="standard"]
rating_probs = ag_hp.groupby('movement_time').size().div(len(ag_hp))
conditional_probs = ag_hp.groupby(['won', 'movement_time']).size().div(len(ag_hp)).div(rating_probs, axis=0, level='movement_time').drop(labels=[-1]).plot(ax=axes[1], style=['b'], label='standard')

ag_hp = df[df['num_honeypots']==0]
ag_hp = ag_hp[ag_hp['agent']=="careful"]
rating_probs = ag_hp.groupby('movement_time').size().div(len(ag_hp))
conditional_probs = ag_hp.groupby(['won', 'movement_time']).size().div(len(ag_hp)).div(rating_probs, axis=0, level='movement_time').drop(labels=[-1])
# add 1 25 with 0
index = pd.MultiIndex.from_tuples([(1, 25)], names=["won", "movement_time"])
s = pd.Series(0.0, index=index)
conditional_probs = pd.concat([s, conditional_probs])
conditional_probs.plot(ax=axes[1], style=['g'], label='careful')

plt.sca(axes[1])
plt.legend()
plt.xlabel("movement_time")
plt.ylabel("probability")
axes[1].title.set_text('b')
plt.xticks(list(range(5)), [25, 50, 75, 100, '-'])


fig.suptitle('Win Probability of Agents HP and MT')
fig.savefig("plots/win_agents_propplot1t.pdf", bbox_inches='tight')
```
